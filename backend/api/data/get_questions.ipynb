```python
with open('egzamin-kierunkowy.txt') as f:
    s = f.read()
```

```python
questions = [q.split('\n') for q in s.split('\n\n')]
```

```python
for q in questions:
    question_tokens = q[0].split(' ')
    q[0] = (question_tokens[1]), ' '.join(question_tokens[2:])
    for i in range(1, len(q)):
        if q[i].startswith(">>"):
            q[i] = q[i].lstrip(">>")
            q[i]= (q[i][3:], True)
        else:
            q[i]= (q[i][3:], False)
            
```

```python
questions
```

```python
questions_json = [{"question": q[0][1], "category": q[0][0], "answers": [{"answer": a[0], "is_correct": a[1]} for a in q[1:]]} for q in questions]
```

```python
import json
```

```python
with open('questions.json', 'w') as f:
    f.write(json.dumps(questions_json, indent=4, ensure_ascii=False))
```

```python
with open('questions.json', 'r') as f:
    print(f.read())
```
